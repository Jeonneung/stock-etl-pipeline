## 데이터 분석

```python
benefit_analyze = stock_info[['종목코드', '종목명', '종가', 'BPS', 'PBR']]
```

```python
benefit_analyze['BPS'] = pd.to_numeric(benefit_analyze['BPS'], errors='coerce')
benefit_analyze['PBR'] = pd.to_numeric(benefit_analyze['PBR'], errors='coerce')
benefit_analyze.info()
```

```python
benefit_analyze['T_Price'] = benefit_analyze['BPS'] * 0.8
benefit_analyze['D_Support'] = benefit_analyze['BPS'] * 0.35
benefit_analyze['U_Potential'] = (benefit_analyze['T_Price'] / benefit_analyze['종가'] - 1) * 100
benefit_analyze['RR_Ratio'] = (benefit_analyze['T_Price'] - benefit_analyze['종가']) / (benefit_analyze['종가'] - benefit_analyze['D_Support'])

benefit_analyze.head()
```

```python
result = benefit_analyze.sort_values(by=['U_Potential'], ascending=False)
```

```python
result.reset_index(drop=True, inplace=True)
```

```python
# 2025 상장폐지 사유 발생 기업 필터링
delisting_stock = ['국보', '웰바이오텍', '세원이엔씨', '아이에이치큐', '이아이디', 'KH 필룩스', '금양', 
                   'KC코트렐', 'KC그린홀딩스', '범양건설', '효성화학', '한국패러랠', '한창']

# result.drop()
result.drop(result[result['종목명'].isin(delisting_stock)].index, inplace=True)
```

# 모듈화 테스트

```python
def get_data_with_api():
    import os
    import requests
    import json

    KRX_API_KEY = os.getenv('KRX_API_KEY')
    # 유가증권 일별매매정보, 코덱스 일별매매정보
    API_ID = ['stk_bydd_trd', 'ksq_bydd_trd']
    SERVER_ENDPOINT_URL = "http://data-dbg.krx.co.kr/svc/apis/sto/"

    targetDate = '20260126'
    params = {'basDd': targetDate}
    headers = {'AUTH_KEY': KRX_API_KEY}

    dfs = []
    for id in API_ID:
        res = requests.get(url=SERVER_ENDPOINT_URL + id, params=params, headers=headers)
        data = json.loads(res.text)

        if 'OutBlock_1' in data:
            data = pd.DataFrame(data['OutBlock_1'])
            dfs.append(data)
        else:
            print(f"No data in {id}")
            continue
        
    result = pd.concat(dfs, axis=0)

    return result
```

```python
api_result = get_data_with_api()
```
